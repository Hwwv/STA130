### Q2: Simplifying the code for readability and efficient.

I ask GPT to simplify the codes above with a simpler, more streamlined way for the loop.

It makes the reveal of Monte easier without adding or deleting not-car-doors by having a list of remaining doors. Then the switch using nesting conditions.

The reason why it is easier to read is that its structure follows the process of game, while constructing a simpler way to do the same action.

Codes-changed are as follows:

```python
import numpy as np

# Number of simulations
reps = 100000
wins = 0  # Counter to track wins

for _ in range(reps):
    # Step 1: Randomly assign the winning door (1, 2, or 3)
    winning_door = np.random.choice([1, 2, 3])
    
    # Step 2: Player makes an initial choice (always choosing door 1 for simplicity)
    player_choice = 1
    
    # Step 3: Monty opens a door with a goat
    # Monty reveals a door that is not the player's choice and not the winning door
    remaining_doors = [door for door in [1, 2, 3] if door != player_choice and door != winning_door]
    monty_reveals = np.random.choice(remaining_doors)
    
    # Step 4: Player switches to the other remaining door
    switch_to = [door for door in [1, 2, 3] if door != player_choice and door != monty_reveals][0]
    
    # Step 5: Check if the switch leads to a win
    if switch_to == winning_door:
        wins += 1

# Calculate and print the win probability when switching
win_probability = wins / reps
print(f"Win probability when switching: {win_probability}")
```

```python
from collections import defaultdict
import random

# Character-specific Markov chains
character_chains = defaultdict(lambda: defaultdict(lambda: defaultdict(int)))

# Function to build a character-specific bigram Markov chain
def build_chain(character, words):
    for i in range(len(words) - 2):
        bigram = (words[i], words[i+1])  # Create bigram from two consecutive words
        next_word = words[i+2]  # The word following the bigram
        character_chains[character][bigram][next_word] += 1  # Increment the count of next_word for the bigram

# Example of building chains for multiple characters
character_dialogues = {
    'Alice': ['hello', 'world', 'hello', 'how', 'are', 'you'],
    'Bob': ['how', 'are', 'you', 'doing', 'today', 'hello', 'world'],
}

for character, words in character_dialogues.items():
    build_chain(character, words)

# Function to generate a response for a character using bigrams
def generate_response(character, start_words, length=10):
    response = list(start_words)  # Start the response with the provided bigram (start words)
    for _ in range(length - 2):  # Generate words up to the desired length
        bigram = (response[-2], response[-1])  # The current bigram is the last two words in the response
        possible_next_words = character_chains[character].get(bigram, None)  # Get next word options for the bigram

        if possible_next_words:
            # Select the next word based on frequency (weighted random choice)
            next_word = random.choices(list(possible_next_words.keys()), list(possible_next_words.values()))[0]
            response.append(next_word)  # Add the selected word to the response
        else:
            break  # If no valid next word is found, stop generating further
    return ' '.join(response)  # Return the generated response as a string

# Example usage: generating responses for Alice and Bob
start_words_alice = ('hello', 'world')
response_alice = generate_response('Alice', start_words_alice)
print("Alice's response:", response_alice)

start_words_bob = ('how', 'are')
response_bob = generate_response('Bob', start_words_bob)
print("Bob's response:", response_bob)
```

### Q2: Using seaborn kernel density estimation\(KDE\) plots to produce plots

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
penguins = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/penguins.csv")

# Group by species and calculate descriptive statistics for flipper_length_mm
stats = penguins.groupby('species')['flipper_length_mm'].agg(
    mean=np.mean,
    median=np.median,
    range_min=lambda x: x.min(),
    range_max=lambda x: x.max(),
    iqr=lambda x: np.percentile(x.dropna(), 75) - np.percentile(x.dropna(), 25),
    std=np.std
).reset_index()

# Add columns for two standard deviations from the mean
stats['lower_2sd'] = stats['mean'] - 2 * stats['std']
stats['upper_2sd'] = stats['mean'] + 2 * stats['std']

# Set up the seaborn style
sns.set(style="whitegrid")

# Create a figure with 1 row and 3 columns
fig, axes = plt.subplots(1, 3, figsize=(18, 5), sharey=True)

species_list = penguins['species'].dropna().unique()

# Plot KDEs with location and scale statistics for each species
for i, species in enumerate(species_list):
    ax = axes[i]
    species_data = penguins[penguins['species'] == species]['flipper_length_mm'].dropna()
    species_stats = stats[stats['species'] == species]
    
    # KDE plot for the species
    sns.kdeplot(species_data, ax=ax, fill=True, color='skyblue', label=f'{species} KDE')
    
    # Add lines for mean, median, and IQR, and shading for ±2 standard deviations
    mean = species_stats['mean'].values[0]
    median = species_stats['median'].values[0]
    lower_2sd = species_stats['lower_2sd'].values[0]
    upper_2sd = species_stats['upper_2sd'].values[0]
    range_min = species_stats['range_min'].values[0]
    range_max = species_stats['range_max'].values[0]
    
    # Vertical lines for mean and median
    ax.axvline(mean, color='green', linestyle='--', label='Mean')
    ax.axvline(median, color='blue', linestyle='--', label='Median')
    
    # Shaded region for ±2 SD
    ax.axvspan(lower_2sd, upper_2sd, color='lightgreen', alpha=0.3, label='±2 SD')
    
    # Shaded region for range
    ax.axvspan(range_min, range_max, color='lightblue', alpha=0.2, label='Range')

    # Add title and labels
    ax.set_title(f'{species} Flipper Length KDE')
    ax.set_xlabel('Flipper Length (mm)')
    ax.set_ylabel('Density')
    
    # Show the legend
    ax.legend()

# Adjust the layout
plt.tight_layout()
plt.show()
```

1. **Mean and Median**:
   - The **mean** is the average of all data points, while the **median** is the middle value when the data is sorted.
   - In a perfectly symmetrical distribution (like a normal distribution), the mean and median are equal.

2. **Right Skewness (Positive Skewness)**:
   - In a right-skewed distribution, the tail on the right side is longer or fatter than the left. 
   - Here, the mean is typically greater than the median because the larger values in the tail pull the mean up. 

3. **Left Skewness (Negative Skewness)**:
   - In a left-skewed distribution, the tail on the left side is longer or fatter than the right.
   - In this case, the mean is usually less than the median because the smaller values in the tail pull the mean down.

I asked the ChatBot to use the coffee dataset for producing a 95% bootstrap confisence interval for a population mean and for population median and population variance.


```python
import numpy as np
import pandas as pd
import requests

# Load the data from the URL
url = "https://raw.githubusercontent.com/rfordatascience/tidytuesday/2e9bd5a67e09b14d01f616b00f7f7e0931515d24/data/2020/2020-07-07/coffee_ratings.csv"
data = pd.read_csv(url)

# Extract the column of interest: 'total_cup_points' (for example)
sample_data = data['total_cup_points'].dropna()  # Drop any missing values

# Number of bootstrap samples
n_bootstraps = 10000

# --- 95% Bootstrap CI for Population Mean ---
bootstrap_means = []

for _ in range(n_bootstraps):
    bootstrap_sample = np.random.choice(sample_data, size=len(sample_data), replace=True)
    bootstrap_mean = np.mean(bootstrap_sample)  # Calculate mean for each bootstrap sample
    bootstrap_means.append(bootstrap_mean)

# Calculate the 95% confidence interval for the mean
lower_bound_mean = np.percentile(bootstrap_means, 2.5)
upper_bound_mean = np.percentile(bootstrap_means, 97.5)

print(f"95% Bootstrapped Confidence Interval for Mean: [{lower_bound_mean:.2f}, {upper_bound_mean:.2f}]")


# --- 95% Bootstrap CI for Population Median ---
bootstrap_medians = []

for _ in range(n_bootstraps):
    bootstrap_sample = np.random.choice(sample_data, size=len(sample_data), replace=True)
    bootstrap_median = np.median(bootstrap_sample)  # Calculate median for each bootstrap sample
    bootstrap_medians.append(bootstrap_median)

# Calculate the 95% confidence interval for the median
lower_bound_median = np.percentile(bootstrap_medians, 2.5)
upper_bound_median = np.percentile(bootstrap_medians, 97.5)

print(f"95% Bootstrapped Confidence Interval for Median: [{lower_bound_median:.2f}, {upper_bound_median:.2f}]")


# --- 95% Bootstrap CI for Population Variance ---
bootstrap_variances = []

for _ in range(n_bootstraps):
    bootstrap_sample = np.random.choice(sample_data, size=len(sample_data), replace=True)
    bootstrap_variance = np.var(bootstrap_sample, ddof=1)  # Calculate sample variance (ddof=1) for each bootstrap sample
    bootstrap_variances.append(bootstrap_variance)

# Calculate the 95% confidence interval for the variance
lower_bound_variance = np.percentile(bootstrap_variances, 2.5)
upper_bound_variance = np.percentile(bootstrap_variances, 97.5)

print(f"95% Bootstrapped Confidence Interval for Variance: [{lower_bound_variance:.2f}, {upper_bound_variance:.2f}]")


# --- 95% Bootstrap CI for Population Standard Deviation ---
bootstrap_std_devs = []

for _ in range(n_bootstraps):
    bootstrap_sample = np.random.choice(sample_data, size=len(sample_data), replace=True)
    bootstrap_std_dev = np.std(bootstrap_sample, ddof=1)  # Calculate sample std deviation (ddof=1) for each bootstrap sample
    bootstrap_std_devs.append(bootstrap_std_dev)

# Calculate the 95% confidence interval for the standard deviation
lower_bound_std_dev = np.percentile(bootstrap_std_devs, 2.5)
upper_bound_std_dev = np.percentile(bootstrap_std_devs, 97.5)

print(f"95% Bootstrapped Confidence Interval for Standard Deviation: [{lower_bound_std_dev:.2f}, {upper_bound_std_dev:.2f}]")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Sample data
data = {
    'PatientID': range(1, 11),
    'Age': [45, 34, 29, 52, 37, 41, 33, 48, 26, 39],
    'Gender': ['M', 'F', 'M', 'F', 'M', 'F', 'M', 'F', 'M', 'F'],
    'InitialHealthScore': [84, 78, 83, 81, 81, 80, 79, 85, 76, 83],
    'FinalHealthScore': [86, 86, 80, 86, 84, 86, 86, 82, 83, 84]
}

df = pd.DataFrame(data)

# Calculate differences
df['Difference'] = df['FinalHealthScore'] - df['InitialHealthScore']
differences = df['Difference'].values

# Bootstrapping parameters
n_iterations = 10000
n_size = len(differences)
bootstrap_means = []

# Perform bootstrapping
for _ in range(n_iterations):
    sample = np.random.choice(differences, size=n_size, replace=True)
    bootstrap_mean = np.mean(sample)
    bootstrap_means.append(bootstrap_mean)

# Convert to numpy array for easier manipulation
bootstrap_means = np.array(bootstrap_means)

# Calculate confidence interval
confidence_level = 0.95
lower_percentile = (1 - confidence_level) / 2 * 100
upper_percentile = (1 + confidence_level) / 2 * 100
conf_interval = np.percentile(bootstrap_means, [lower_percentile, upper_percentile])

# Calculate the proportion of bootstrap means greater than 0
prop_greater_than_zero = np.mean(bootstrap_means > 0)

# Output results
print(f"Mean of original differences: {np.mean(differences):.2f}")
print(f"95% Confidence Interval for the mean difference: {conf_interval[0]:.2f} to {conf_interval[1]:.2f}")
print(f"Proportion of bootstrap means > 0: {prop_greater_than_zero:.4f}")

# Plot the bootstrap distribution
plt.hist(bootstrap_means, bins=50, edgecolor='k', alpha=0.7)
plt.axvline(x=0, color='red', linestyle='--', label='Difference = 0')
plt.axvline(x=conf_interval[0], color='blue', linestyle='--', label='95% CI Lower')
plt.axvline(x=conf_interval[1], color='blue', linestyle='--', label='95% CI Upper')
plt.title('Bootstrap Distribution of Mean Differences')
plt.xlabel('Mean Difference')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
# Import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# -------------------------------
# 1. Data Preparation
# -------------------------------

# Define the dataset
data = {
    'PatientID': range(1, 11),
    'Age': [45, 34, 29, 52, 37, 41, 33, 48, 26, 39],
    'Gender': ['M', 'F', 'M', 'F', 'M', 'F', 'M', 'F', 'M', 'F'],
    'InitialHealthScore': [84, 78, 83, 81, 81, 80, 79, 85, 76, 83],
    'FinalHealthScore': [86, 86, 80, 86, 84, 86, 86, 82, 83, 84]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Calculate the difference between Final and Initial Health Scores
df['Difference'] = df['FinalHealthScore'] - df['InitialHealthScore']
differences = df['Difference'].values

# Display the dataset with differences
print("Dataset with Differences:")
print(df)

# -------------------------------
# 2. One-Sided Bootstrapping
# -------------------------------

# Set random seed for reproducibility
np.random.seed(42)

# Bootstrapping parameters
n_iterations = 10000  # Number of bootstrap samples
n_size = len(differences)  # Sample size

# Initialize an array to store bootstrap means
bootstrap_means = np.empty(n_iterations)

# Perform bootstrapping
for i in range(n_iterations):
    # Resample with replacement
    sample = np.random.choice(differences, size=n_size, replace=True)
    # Compute the mean of the resample
    bootstrap_means[i] = np.mean(sample)

# Calculate the observed mean difference
observed_mean = np.mean(differences)

# Calculate the 95% confidence lower bound for the mean difference
confidence_level = 0.95
lower_percentile = (1 - confidence_level) * 100  # 5th percentile
conf_interval_lower = np.percentile(bootstrap_means, lower_percentile)

# Calculate the proportion of bootstrap means greater than 0 (p-value)
prop_greater_than_zero = np.mean(bootstrap_means > 0)

# Output bootstrapping results
print(f"\nOne-Sided Hypothesis Testing Results (Bootstrapping):")
print(f"Mean of original differences: {observed_mean:.2f}")
print(f"95% Confidence Lower Bound for the mean difference: {conf_interval_lower:.2f}")
print(f"Proportion of bootstrap means > 0 (p-value): {prop_greater_than_zero:.4f}")

# -------------------------------
# 3. One-Sided Paired t-Test
# -------------------------------

# Perform one-sided paired t-test
# Null Hypothesis: mean difference <= 0
# Alternative Hypothesis: mean difference > 0

# Calculate t-statistic and two-tailed p-value
t_stat, p_val_two_tailed = stats.ttest_1samp(differences, 0)

# Convert two-tailed p-value to one-tailed p-value
if t_stat > 0:
    p_val_one_tailed = p_val_two_tailed / 2
else:
    p_val_one_tailed = 1 - (p_val_two_tailed / 2)

# Output t-test results
print(f"\nOne-Sided Paired t-Test Results:")
print(f"t-statistic: {t_stat:.3f}")
print(f"One-tailed p-value: {p_val_one_tailed:.4f}")

# -------------------------------
# 4. Effect Size Calculation
# -------------------------------

# Calculate Cohen's d for effect size
cohen_d = observed_mean / np.std(differences, ddof=1)
print(f"\nEffect Size (Cohen's d): {cohen_d:.3f}")

# -------------------------------
# 5. Visualization of Results
# -------------------------------

# a. Bootstrap Distribution Plot
plt.figure(figsize=(10, 6))
sns.histplot(bootstrap_means, bins=50, kde=True, color='skyblue', edgecolor='k')
plt.axvline(x=0, color='red', linestyle='--', label='Difference = 0')
plt.axvline(x=conf_interval_lower, color='blue', linestyle='--', label='95% CI Lower')
plt.axvline(x=observed_mean, color='green', linestyle='-', label='Observed Mean')
plt.title('Bootstrap Distribution of Mean Differences (One-Sided Test)')
plt.xlabel('Mean Difference')
plt.ylabel('Frequency')
plt.legend()
plt.show()

# b. One-Sided t-Test Visualization
# Degrees of freedom
df_t = len(differences) - 1

# Critical t-value for one-tailed test at alpha=0.05
alpha = 0.05
t_critical = stats.t.ppf(1 - alpha, df=df_t)

# Generate t-distribution values
t_values = np.linspace(-4, 4, 1000)
t_dist = stats.t.pdf(t_values, df=df_t)

# Plot the t-distribution with critical region and t-statistic
plt.figure(figsize=(10, 6))
plt.plot(t_values, t_dist, label='t-Distribution', color='purple')
plt.fill_between(t_values, t_dist, where=(t_values >= t_critical), color='salmon', alpha=0.5, label='Critical Region (α=0.05)')
plt.axvline(x=t_stat, color='blue', linestyle='--', label=f't-statistic = {t_stat:.3f}')
plt.title('One-Sided t-Test Visualization')
plt.xlabel('t-value')
plt.ylabel('Probability Density')
plt.legend()
plt.show()

# -------------------------------
# 6. Confidence Interval via Bootstrapping (Optional)
# -------------------------------

# Calculate 95% confidence interval using bootstrapping
# For one-sided, we're primarily interested in the lower bound
ci_lower = conf_interval_lower
ci_upper = np.percentile(bootstrap_means, 100 - lower_percentile)

print(f"\n95% Confidence Interval for Mean Difference (Bootstrapping): {ci_lower:.2f} to {ci_upper:.2f}")

# -------------------------------
# 7. Final Summary
# -------------------------------

print("\n--- Summary of Findings ---")
print(f"Observed Mean Difference: {observed_mean:.2f}")
print(f"95% Confidence Lower Bound: {ci_lower:.2f}")
print(f"95% Confidence Interval: {ci_lower:.2f} to {ci_upper:.2f}")
print(f"Proportion of bootstrap means > 0 (p-value): {prop_greater_than_zero:.4f}")
print(f"t-statistic: {t_stat:.3f}")
print(f"One-tailed p-value: {p_val_one_tailed:.4f}")
print(f"Cohen's d (Effect Size): {cohen_d:.3f}")

if prop_greater_than_zero > (1 - alpha):
    print("\nConclusion: Reject the null hypothesis. There is strong evidence that the vaccine significantly improves health scores.")
else:
    print("\nConclusion: Fail to reject the null hypothesis. There is insufficient evidence to conclude that the vaccine improves health scores.")
```

Given the experiment from Fisher and Bristol, I want to see the result that Fisher got is related to the personal identity of Bristol, which made the conclusion full of individual 'error', rather than the difference between milk before tea or tea before milk itself.

Now we use the sample from class for the combination of random observations to reduce the individual's impact on the general result.

Now, let's say, if a student tastes the difference, then we add one point to the difference; by contrast, if not, we add one point to the non-difference. Then we finally have the percentage of each situation. The $H_0$ here is there is no difference about the pouring order, that is $p_1$=$p_2$, the first is the difference-group as the second for the non-difference group.
```python
# Import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# -------------------------------
# 1. Data Preparation
# -------------------------------

# Set random seed for reproducibility
np.random.seed(42)

# Sample data
data = {
    'StudentID': range(1, 81),  # 80 students
    'CorrectlyIdentified': [1]*49 + [0]*(80-49)  # 49 correct, 31 incorrect
}

# Create DataFrame
df = pd.DataFrame(data)

# Shuffle the DataFrame to randomize correct identifications among students
df = df.sample(frac=1, random_state=42).reset_index(drop=True)

# Display the first few rows
print("Dataset Sample:")
print(df.head())

# -------------------------------
# 2. One-Sided Hypothesis Test Using Z-Test
# -------------------------------

# Total number of students
n = len(df)

# Number of correct identifications
x = df['CorrectlyIdentified'].sum()

# Observed proportion
p_hat = x / n

# Hypothesized proportion under H0
p0 = 0.5

# Standard error
SE = np.sqrt(p0 * (1 - p0) / n)

# Z-statistic
z = (p_hat - p0) / SE

# One-tailed p-value
p_value = 1 - stats.norm.cdf(z)

# Output the results
print(f"\nOne-Sample Proportion Z-Test Results:")
print(f"Observed Proportion (p̂): {p_hat:.3f}")
print(f"Null Hypothesis Proportion (p0): {p0:.3f}")
print(f"Z-Statistic: {z:.3f}")
print(f"One-Tailed P-Value: {p_value:.4f}")

# -------------------------------
# 3. Bootstrapping Approach
# -------------------------------

# Number of bootstrap samples
n_iterations = 10000

# Initialize an array to store bootstrap proportions
bootstrap_props = np.empty(n_iterations)

# Perform bootstrapping
for i in range(n_iterations):
    # Generate a bootstrap sample under H0 (p = 0.5)
    sample = np.random.binomial(1, p0, n)
    # Calculate the proportion of correct identifications in the sample
    bootstrap_props[i] = sample.mean()

# Calculate the p-value: proportion of bootstrap samples >= observed proportion
bootstrap_p_value = np.mean(bootstrap_props >= p_hat)

# Output the bootstrap p-value
print(f"\nBootstrapping Results:")
print(f"Bootstrap P-Value: {bootstrap_p_value:.4f}")

# -------------------------------
# 4. Visualization of Bootstrap Distribution
# -------------------------------

# Plot the bootstrap distribution
plt.figure(figsize=(10, 6))
sns.histplot(bootstrap_props, bins=50, kde=True, color='skyblue', edgecolor='k')
plt.axvline(x=p_hat, color='red', linestyle='--', label='Observed Proportion')
plt.axvline(x=0.5, color='green', linestyle='-', label='Null Hypothesis Proportion (p=0.5)')
plt.title('Bootstrap Distribution of Proportion under H₀')
plt.xlabel('Proportion of Correct Identifications')
plt.ylabel('Frequency')
plt.legend()
plt.show()

# -------------------------------
# 5. Summary of Findings
# -------------------------------

print("\n--- Summary of Findings ---")
print(f"Observed Proportion (p̂): {p_hat:.3f}")
print(f"Null Hypothesis Proportion (p0): {p0:.3f}")
print(f"Z-Statistic: {z:.3f}")
print(f"One-Tailed P-Value (Z-Test): {p_value:.4f}")
print(f"Bootstrap P-Value: {bootstrap_p_value:.4f}")

if p_value < 0.05:
    print("\nConclusion: Reject the null hypothesis. There is significant evidence that the proportion of correct identifications is greater than chance.")
else:
    print("\nConclusion: Fail to reject the null hypothesis. There is insufficient evidence to conclude that the proportion of correct identifications is greater than chance.")
```

### Q1: Explanation on theoretical Simple Linear Regression model by describing the components and how they combine to form a sample from normal distribution; then, create python code explicitly demonstrating your explanation using numpy and scipy.stats

Simple linear regression model in theory, as I noticed in the beginning part, is used for catching the possible relation between variables. In the simplest case, we have two variables and we want to get the relation between them. Then we must guess one variable's changing could affect another. The "input", or say, self-changing variable is the **predictor variable**. By the contrast, the affected one is an **outcome variable**. The effect of changing one unit of "predictor" on the "outcome" is the slope if the whold model is linear. And the intercept coefficient is the value of outcome variable when the predictor variable is zero, which means this part of outcome will not be affected by the change of "input". However, we can not say that our prediction is 100% correct, so there is the error term in the model for the part we are not able to describe in the relation, like uncertainty.

To get the regression model, we can not simply just use a few points. Generally speaking, every pair of predictor and outcome and selection are following the normal distribution. We need to get a lot of samples(pairs) randomly following normal distribution to get a general model of the two variables. So every selection get one possible regression, all the selections' regressions finally will turn out to be normal distribution.

```python
import numpy as np
import plotly.graph_objects as go
from scipy.stats import norm

# Set random seed for reproducibility
np.random.seed(42)

# Define parameters
n = 100              # Number of data points
beta_0 = 2           # Intercept
beta_1 = 3           # Slope
sigma = 1            # Standard deviation of errors

# Generate random X values (can be uniformly distributed)
X = np.random.uniform(0, 10, n)

# Generate random error terms from a normal distribution
errors = norm.rvs(loc=0, scale=sigma, size=n)

# Calculate Y using the linear model: Y = beta_0 + beta_1 * X + errors
Y = beta_0 + beta_1 * X + errors

# Create the plot with Plotly
fig = go.Figure()

# Scatter plot of the generated data
fig.add_trace(go.Scatter(x=X, y=Y, mode='markers', name='Simulated Data', marker=dict(color='blue')))

# Plot the true regression line (without noise)
fig.add_trace(go.Scatter(x=X, y=beta_0 + beta_1 * X, mode='lines', name='True Regression Line', line=dict(color='red')))

# Customize the layout
fig.update_layout(
    title='Simulated Data and True Regression Line',
    xaxis_title='X',
    yaxis_title='Y',
    template='plotly_white',
    legend=dict(x=0.8, y=1),
    width=800,
    height=500
)

# Show the plot
fig.show(renderer = 'png')
```

### Q13: The "short" and "long" wait times are not "before and after" measurements so there are not natural pairs on which to base differences on which to do a "one sample" (paired differences) hypothesis test; but, we can do "two sample" hypothesis testing using a permuation test, or create a 95% bootstrap confidence interval for the difference in means of the two populations.
1. Do a permuation test $H_0: \mu_short = \mu_long$ no difference in duration between short and long groups by "shuffling" the labels
2. Create a 95% bootstrap confidence interval by repeatedly bootstrapping within each group and applying np.quantile(bootstrapped_mean_differences, \[0.025, 0.975\]) to the collection of differences between the sample means.
3. Explain how the sampling approaches work for the two simulations.
4. Compare and contrast these two methods with the indicator variable based model approach used in Question 11, explaining how they're similar and different.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import plotly.graph_objects as go

# Load the Old Faithful geyser dataset
old_faithful = sns.load_dataset('geyser')

# Create the 'kind' indicator variable based on a 68-minute threshold
old_faithful['kind'] = ['short' if wait < 68 else 'long' for wait in old_faithful['waiting']]

# Separate durations into short and long groups
short_durations = old_faithful[old_faithful['kind'] == 'short']['duration']
long_durations = old_faithful[old_faithful['kind'] == 'long']['duration']

# Calculate observed difference in means
observed_diff = short_durations.mean() - long_durations.mean()
print(f"Observed Difference in Means: {observed_diff}")

# Define the permutation test function
def permutation_test(short, long, num_permutations=10000):
    combined = np.concatenate([short, long])
    n_short = len(short)
    n_long = len(long)
    
    # Store differences
    permuted_diffs = []
    
    for _ in range(num_permutations):
        # Shuffle combined array
        np.random.shuffle(combined)
        
        # Split back into two groups
        new_short = combined[:n_short]
        new_long = combined[n_short:]
        
        # Calculate difference in means
        permuted_diff = new_short.mean() - new_long.mean()
        permuted_diffs.append(permuted_diff)
    
    return np.array(permuted_diffs)

# Run the permutation test
permuted_differences = permutation_test(short_durations.values, long_durations.values)

# Calculate p-value
p_value = np.mean(np.abs(permuted_differences) >= np.abs(observed_diff))
print(f"P-value from Permutation Test: {p_value}")

# Visualize the permutation distribution using Plotly
fig = go.Figure()

# Add histogram of permuted differences
fig.add_trace(go.Histogram(
    x=permuted_differences,
    nbinsx=30,
    name='Permuted Differences',
    marker_color='lightblue',
    opacity=0.7
))

# Add a line for the observed difference
fig.add_trace(go.Scatter(
    x=[observed_diff, observed_diff],
    y=[0, 2000],  # Adjust y range according to your histogram
    mode='lines',
    line=dict(color='red', width=2, dash='dash'),
    name='Observed Difference'
))

# Update layout
fig.update_layout(
    title='Permutation Test: Distribution of Differences in Means',
    xaxis_title='Difference in Means',
    yaxis_title='Frequency',
    showlegend=True
)

# Show the figure
fig.show(renderer = "png")
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import plotly.graph_objects as go

# Load the Old Faithful geyser dataset
old_faithful = sns.load_dataset('geyser')

# Create the 'kind' indicator variable based on a 68-minute threshold
old_faithful['kind'] = ['short' if wait < 68 else 'long' for wait in old_faithful['waiting']]

# Separate durations into short and long groups
short_durations = old_faithful[old_faithful['kind'] == 'short']['duration'].values
long_durations = old_faithful[old_faithful['kind'] == 'long']['duration'].values

# Bootstrap function to calculate differences in means
def bootstrap_difference_in_means(short, long, num_bootstraps=10000):
    bootstrapped_mean_diffs = []
    
    for _ in range(num_bootstraps):
        # Bootstrap sample for short and long groups
        short_sample = np.random.choice(short, size=len(short), replace=True)
        long_sample = np.random.choice(long, size=len(long), replace=True)
        
        # Calculate mean difference
        mean_diff = short_sample.mean() - long_sample.mean()
        bootstrapped_mean_diffs.append(mean_diff)
    
    return np.array(bootstrapped_mean_diffs)

# Perform bootstrap
bootstrapped_mean_differences = bootstrap_difference_in_means(short_durations, long_durations)

# Calculate 95% bootstrap confidence interval
conf_interval = np.quantile(bootstrapped_mean_differences, [0.025, 0.975])
print(f"95% Bootstrap Confidence Interval for the Difference in Means: {conf_interval}")

# Visualize the bootstrap distribution using Plotly
fig = go.Figure()

# Add histogram of bootstrapped mean differences
fig.add_trace(go.Histogram(
    x=bootstrapped_mean_differences,
    nbinsx=30,
    name='Bootstrapped Mean Differences',
    marker_color='lightblue',
    opacity=0.7
))

# Add lines for the confidence interval
fig.add_trace(go.Scatter(
    x=[conf_interval[0], conf_interval[0]],
    y=[0, 2000],  # Adjust y range according to your histogram
    mode='lines',
    line=dict(color='green', width=2, dash='dash'),
    name='Lower CI (2.5%)'
))

fig.add_trace(go.Scatter(
    x=[conf_interval[1], conf_interval[1]],
    y=[0, 2000],  # Adjust y range according to your histogram
    mode='lines',
    line=dict(color='red', width=2, dash='dash'),
    name='Upper CI (97.5%)'
))

# Update layout
fig.update_layout(
    title='Bootstrap Confidence Interval for the Difference in Means',
    xaxis_title='Difference in Means',
    yaxis_title='Frequency',
    showlegend=True
)

# Show the figure
fig.show(renderer = "png")
```

Key Differences Between the Two Approaches
Purpose:<br>
Permutation Test: Primarily a hypothesis test focused on assessing whether the observed difference in means is statistically significant.<br>
Bootstrap Confidence Interval: Focused on estimating the uncertainty around the difference in means and quantifying that uncertainty via a confidence interval.<br>
Data Handling:<br>
Permutation Test: Shuffles data while maintaining the overall structure to create a distribution under the null hypothesis.

Bootstrap: Resamples with replacement from the actual data, which allows for capturing variability and estimating the sampling distribution directly from the observed data.

Result Interpretation:

Permutation Test: The p-value indicates the likelihood of observing such a difference if there were no true effect.

Bootstrap Confidence Interval: The confidence interval provides a range of plausible values for the true mean difference, indicating the precision of the estimate.
